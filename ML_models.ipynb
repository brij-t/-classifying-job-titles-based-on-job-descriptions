```python
import numpy as nd
import pandas as pd
import sklearn
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import confusion_matrix, classification_report,accuracy_score
from sklearn.pipeline import Pipeline
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC  
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.ensemble import VotingClassifier
```

```python
data = pd.read_csv('Top30.csv')
```

```python
data.head()
```

```python
data = data.drop(["Unnamed: 0","Unnamed: 0.1"], axis=1)
```

```python
data.dtypes
```

```python
print(data.Query.value_counts())
```

```python
#Split the dataset to Training and Test subsets (90/10)
from sklearn.model_selection import train_test_split
train, test = train_test_split(data, test_size = 0.2, random_state = 17) #random_state = None

train_descs = train['Description']
train_labels = train['Query']
 
test_descs = test['Description']
test_labels = test['Query']


X_train, X_val, y_train, y_val  = train_test_split(train_descs, train_labels, test_size=0.2, random_state=17)
```

```python
print(len(train_descs),len(train_labels),len(test_descs),len(test_labels),len(X_train),len(X_val),len(y_train),len(y_val))
```

```python
###Naive bayes
text_clf = Pipeline([('vect', CountVectorizer()), ('tfidf', TfidfTransformer()), ('clf', MultinomialNB())])
text_clf = text_clf.fit(train_descs, train_labels)
```

```python
predicted = text_clf.predict(test_descs)
np.mean(predicted == test_labels)
print('Accuracy\n',accuracy_score(predicted, test_labels))
print('Confusion Matrix\n',confusion_matrix(predicted, test_labels))
print('classification_score\n',classification_report(predicted, test_labels))
```

```python
###Randomforest
text_rfc = Pipeline([('vect', CountVectorizer()), ('tfidf', TfidfTransformer()), ('rfc', RandomForestClassifier())])
text_rfc = text_rfc.fit(train_descs, train_labels)
```

```python
predicted = text_rfc.predict(test_descs)
np.mean(predicted == test_labels)
print('Accuracy\n',accuracy_score(predicted, test_labels))
print('Confusion Matrix\n',confusion_matrix(predicted, test_labels))
print('classification_score\n',classification_report(predicted, test_labels))
```

```python
###Gradientboost
text_GB = Pipeline([('vect', CountVectorizer()), ('tfidf', TfidfTransformer()), ('Gradient', GradientBoostingClassifier(n_estimators=20, max_features=2, max_depth=2,random_state=0))])
text_GB = text_GB.fit(train_descs, train_labels)
```

```python
predicted = text_GB.predict(test_descs)
np.mean(predicted == test_labels)
print('Accuracy\n',accuracy_score(predicted, test_labels))
print('Confusion Matrix\n',confusion_matrix(predicted, test_labels))
print('classification_score\n',classification_report(predicted, test_labels))
```

```python
###DecisionTree
text_DT = Pipeline([('vect', CountVectorizer()), ('tfidf', TfidfTransformer()), ('DT', DecisionTreeClassifier())])
text_DT = text_DT.fit(train_descs, train_labels)
```

```python
predicted = text_DT.predict(test_descs)
np.mean(predicted == test_labels)
print('Accuracy\n',accuracy_score(predicted, test_labels))
print('Confusion Matrix\n',confusion_matrix(predicted, test_labels))
print('classification_score\n',classification_report(predicted, test_labels))
```

```python
###SVM
text_SVM = Pipeline([('vect', CountVectorizer()), ('tfidf', TfidfTransformer()), ('SVM',SVC(kernel='linear',probability=True))])
text_SVM = text_SVM.fit(train_descs, train_labels)
```

```python
predicted = text_SVM.predict(test_descs)
np.mean(predicted == test_labels)
print('Accuracy\n',accuracy_score(predicted, test_labels))
print('Confusion Matrix\n',confusion_matrix(predicted, test_labels))
print('classification_score\n',classification_report(predicted, test_labels))
```

```python
##Adaboost
text_AD = Pipeline([('vect', CountVectorizer()), ('tfidf', TfidfTransformer()), ('AD', AdaBoostClassifier(DecisionTreeClassifier(max_depth=1)))])
text_AD = text_AD.fit(train_descs, train_labels)
```

```python
predicted = text_AD.predict(test_descs)
np.mean(predicted == test_labels)
print('Accuracy\n',accuracy_score(predicted, test_labels))
print('Confusion Matrix\n',confusion_matrix(predicted, test_labels))
print('classification_score\n',classification_report(predicted, test_labels))
```

```python
##Logisticregresasion
text_LR = Pipeline([('vect', CountVectorizer()), ('tfidf', TfidfTransformer()), ('LR', LogisticRegression(random_state=0))])
text_LR = text_LR.fit(train_descs, train_labels)
```

```python
predicted = text_LR.predict(test_descs)
np.mean(predicted == test_labels)
print('Accuracy\n',accuracy_score(predicted, test_labels))
print('Confusion Matrix\n',confusion_matrix(predicted, test_labels))
print('classification_score\n',classification_report(predicted, test_labels))
```

```python
voting_clf = VotingClassifier(
estimators = [('lr', text_LR), ('svc',text_SVM),('DT',text_DT),('rf', text_rfc),('NB',text_clf)],
voting = 'soft',weights = [3,3,2.5,2,1])
voting_clf.fit(train_descs, train_labels)
```

```python
x_pred = voting_clf.predict(test_descs)
np.mean(x_pred == test_labels)
print('Accuracy\n',accuracy_score(predicted, test_labels))
print('classification_score\n',classification_report(predicted, test_labels))
```
